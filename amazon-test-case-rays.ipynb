# Задача с собеседования в Amazon

Сама задача была подсмотрена в блоге https://pycoder.ru/azamon-interview-tasks/

В задаче нужно определить минимальное количество лучей, выпущенных из начала координат, и проходящих через все точки на плоскости. 



На рисунке у нас во втором квадранте две точки, и по одной точке в остальных. Следовательно, нам понадобится 5 лучей. Чтобы узнать минимальное количество лучей можно рассчитать, проходят ли они через несколько точек одновременно.

Для задачи задается класс `Point2D`:

```Python
class Point2D(object):

    def __init__(self, x, y):
        self.x = x
        self.y = y
```

Примеры того, как функция должна работать, на вход задается массив точек:

```
task2([Point2D(1, 1), Point2D(1, 2)]) -> 2
task2([Point2D(1, 1), Point2D(2, 2), Point2D(3, 3)]) -> 1
```

## Решение

````python
import math
import matplotlib.pyplot as plt
from typing import List, Set
````

````python
class Point2D:

    def __init__(self, x, y):
        self.x = x
        self.y = y
````

````python
def unique_rays(points: List[Point2D]) -> Set[float]:
    radians = []
    for point in points:
        point_1 = Point2D(point.x, 0) # проекция точки на ось X
        # вычисляем угол в радианах между вектором, образованным началом координат и точкой, и осью абсцисс
        # Math.Atan2(vector2.Y, vector2.X) - Math.Atan2(vector1.Y, vector1.X)
        radians.append(math.atan2(point.y, point.x) - math.atan2(point_1.y, point_1.x))
        # возвращаем количество уникальных значений
    return len(set(radians))
````

Проверяем

````python
assert unique_rays([Point2D(1, 1), Point2D(1, 2)]) == 2
assert unique_rays([Point2D(1, 1), Point2D(2, 2), Point2D(3, 3)]) == 1
assert unique_rays([Point2D(1, 1), Point2D(2, 2), Point2D(3, 3), Point2D(-1, 2)]) == 2
assert unique_rays([Point2D(3, 2), Point2D(1, -1), Point2D(3, -1), Point2D(-2, -4), Point2D(-2, 3)]) == 5
````

````python
points = [Point2D(3, 2), Point2D(1, -1), Point2D(3, -1), Point2D(-2, -4), Point2D(-2, 3)]

fig, ax = plt.subplots(figsize=(8,6))
ax.set(frame_on=False)
plt.axhline(y=0.0, color='k', linestyle='-', linewidth=.8)
plt.axvline(x=0.0, color='k', linestyle='-', linewidth=.8)
for n, point in enumerate(points):
    plt.plot([0, point.x], [0, point.y], 'o-', alpha=.5, label=f'Луч {n+1}')
plt.legend();
````
